```python
import numpy as np
import matplotlib.pyplot
import pandas as pd
import seaborn as sns
```

```python
df=pd.DataFrame(dict(time=np.arange(500),value=np.random.randn(500).cumsum()))
new=sns.relplot(x="time",y="value",kind="line",data=df)
new.fig.autofmt_xdate()
```
