```python
import numpy as np
import pandas as pd
x=10**np.arange(1,10)
y=x*2
data=pd.DataFrame(data={'x':x,'y':y})
#lmplot
lmgrid=sns.lmplot(x='x',y='y',data=data,size=7,scatter_kws={"s":100})
lmgrid.set_xticklabels(rotation=90)
plt.show()
```
