```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style="darkgrid") 
```

```python
tips=sns.load_dataset("tips")
print(tips)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=tips,style="smoker")
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue="size")
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",data=tips,palette="ch:r=-.5,l=.75")
#hue is for color,style
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",data=tips,palette="ch:r=-.5,l=.75")
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",size="size",data=tips)
#size is for giving size accor to data hue is for one color in gradient form
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",size="size",sizes=(15,80),data=tips)
```
