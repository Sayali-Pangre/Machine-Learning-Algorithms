```python
#sin plot
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,2*np.pi,100)
y=np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,2*np.pi,100)
Ya=np.sin(x)
Yb=np.cos(x)

plt.plot(x,Ya)
plt.plot(x,Yb)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data=[[5.,25.,50.,20.],
     [4.,23.,51.,17.],
     [6.,22.,52.,19.]]
x=np.arange(4)
plt.bar(x+0.00,data[0],color='r',width=0.25)
plt.bar(x+0.25,data[1],color='g',width=0.25)
plt.bar(x+0.50,data[2],color='b',width=0.25)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
A=[5.,30.,45.,22.]
B=[5.,25.,50.,20.]
x=range(4)
plt.bar(x,A,color='r')
plt.bar(x,B,color='b',bottom=A)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
A=np.array([5.,30.,45.,22.])
B=np.array([5.,25.,50.,20.])
C=np.array([1.,2.,3.,4.])
x=range(4)
plt.bar(x,A,color='r')
plt.bar(x,B,color='b',bottom=A)
plt.bar(x,C,color='g',bottom=A+B)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

female_pop=np.array([5.,30.,45.,22.])
male_pop=np.array([5.,25.,50.,20.])
x=np.arange(4)
plt.barh(x,female_pop,color='r')
plt.barh(x,-male_pop,color='b')
plt.show()
```
